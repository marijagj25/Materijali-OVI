```python
N = 4
initial_state = (((0,)*N),)*N
initial_state
```

#namesto senzori oznaceni so A B C D, kje koristam obelezhani so 1 2 3 4. Praznoto pole e obelezhano so 0.

```python
def is_valid(state, sensor_type, sensor_position):
    row, col = sensor_position
    if sensor_type in state[row]:
        return False

    for i in range(N): #proveruvame dali senzorot od tip senzor_type vekje se naogja vo kolonata kade sto sakame da go vmetneme 
        if state[i][col] == sensor_type:
            return False
    return True
```

```python
state = (
    (0, 1, 2, 3),
    (0, 0, 0, 2),
    (2, 3, 1, 0),
    (0, 0, 0, 0)
)
is_valid(state, 4,(1,0))
```

```python
def add_senzor(state, sensor_type, sensor_position):
    row, col = sensor_position
    state_list = list(list(row) for row in state) #state se pravi kako lista od listi, namesto torki, za da mozheeme da gi menuvame elementite
    state_list[row][col] = sensor_type
    return tuple(tuple(row) for row in state_list) #ja vrakjame state ama povtorno kako torka od torki
```

```python
add_senzor(state, 4,(0,0))
```

```python
def expand_state(state):
    states = []
    for row_index, row in enumerate(state):
        for col_index, element in enumerate(row):
            if element == 0: #toa znaci deka nema senzor na taa pozicija, pa pozicijata mozemee da ja popolnime so senzor
                for sensor_type in range(1, 5):
                    sensor_position = row_index, col_index
                    if is_valid(state, sensor_type, sensor_position):
                        states.append(add_senzor(state, sensor_type, sensor_position))
    return states
```

```python
def depth_first_search_find_path(initial_state):
    queue = deque([initial_state])
    visited = {initial_state}

    while queue:
        state_to_expand = queue.popleft()

        for next_state in expand_state(state_to_expand):
            if next_state not in visited:
                how_many_zeroes = 0
                for row in next_state:
                    how_many_zeroes += row.count(0)

                if how_many_zeroes == 0:
                    return next_state

                queue.appendleft(next_state)
                visited.add(next_state)
```
