Задача 2.10

(10 поени) Имплементирај уште еден алгоритам по твој избор, кој ќе го предвидува квалитетот на вината. Толкувај дали тој алгоритам е подобар од алгоритмот за дрва за одлучување.

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
label_encoder_ph = LabelEncoder()
label_encoder_alcohol = LabelEncoder()

df['pH'] = label_encoder_ph.fit_transform(df['pH'])
df['alcohol'] = label_encoder_alcohol.fit_transform(df['alcohol'])

x = df.iloc[:, :-2]
y = df['quality']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=92
)

mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=92)
mlp.fit(x_train, y_train)
```

```python
predictions = mlp.predict(x_test)
predictions
```

```python
accuracy = accuracy_score(y_test, predictions)
accuracy
```

```python
Според добиените резултати можеме да заклучиме дека DecisionTreeClassifier подобро го предвидува (погодува) квалитетот на вината отколку MLPClassifier.
```
