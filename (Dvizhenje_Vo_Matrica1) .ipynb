1. A Sostojbata za ovoj problem kje ja definiram kako ureden par od dve koordinati:
- pozicija na agentot i 
- pozicija na kutijata.
Odnosno:
state = ((ax, ay), (bx, by))
kade (ax, ay) se koordinatite (red i kolona) na agentot, a (bx, by) se koordinatite na kutijata.
Ova definicija ja izbrav bidejki tablata e fiksna shto ne se menuva,a vo tekot na prebaruvanjeto se menuvaat samo polozhbata na agentot i kutiajta.
Dopolnitelno, za ponatamu da mozham da ja presmetam potroshenata energija, moram da znam kade se naogja agentot, a ne samo kutijata.

```python
def expand_state(state):
    (ax, ay), (bx, by) = state
    next_states = []

    for dx, dy in DIRECTIONS:
        nax = ax + dx
        nay = ay + dy
        # –ê–∫–æ –∞–≥–µ–Ω—Ç–æ—Ç –æ–¥–∏ –Ω–∞–¥–≤–æ—Ä –æ–¥ —Ç–∞–±–ª–∞—Ç–∞ -> –≥–æ –∏–≥–Ω–æ—Ä–∏—Ä–∞–º–µ —Ç–æ—ò –ø–æ—Ç–µ–≥
        if not in_bounds(nax, nay):
            continue
        # 1) –î–≤–∏–∂–µ—ö–µ –ë–ï–ó —Ç—É—Ä–∫–∞—ö–µ –∫—É—Ç–∏—ò–∞
        if (nax, nay) != (bx, by):
            # –∞–≥–µ–Ω—Ç–æ—Ç –æ–¥–∏ –Ω–∞ (nax, nay), –∫—É—Ç–∏—ò–∞—Ç–∞ –æ—Å—Ç–∞–Ω—É–≤–∞ –Ω–∞ –∏—Å—Ç–æ –º–µ—Å—Ç–æ
            new_agent_pos = (nax, nay)
            new_box_pos = (bx, by)
            new_state = (new_agent_pos, new_box_pos)
            next_states.append(new_state)
        # 2) –î–≤–∏–∂–µ—ö–µ –°–û —Ç—É—Ä–∫–∞—ö–µ –∫—É—Ç–∏—ò–∞
        else:
            # –ê–∫–æ –∞–≥–µ–Ω—Ç–æ—Ç —Å–∞–∫–∞ –¥–∞ –≤–ª–µ–∑–µ –≤–æ –∫—É—Ç–∏—ò–∞—Ç–∞, —ò–∞ —Ç—É—Ä–∫–∞ –∫—É—Ç–∏—ò–∞—Ç–∞ –ø–æ–Ω–∞—Ç–∞–º—É
            nbx = bx + dx
            nby = by + dy

            # –ö—É—Ç–∏—ò–∞—Ç–∞ –º–æ—Ä–∞ –∏—Å—Ç–æ –¥–∞ –æ—Å—Ç–∞–Ω–µ –≤–æ –≥—Ä–∞–Ω–∏—Ü–∏
            if not in_bounds(nbx, nby):
                continue
            # –ê–∫–æ –∏–º–∞—Ç–µ —ï–∏–¥–æ–≤–∏, —Ç—É–∫–∞ –±–∏ –ø—Ä–æ–≤–µ—Ä–∏–ª–µ –¥–∞–ª–∏ (nbx, nby) –µ —ï–∏–¥.
            # –ù–∏–µ –ø—Ä–µ—Ç–ø–æ—Å—Ç–∞–≤—É–≤–∞–º–µ –¥–µ–∫–∞ –Ω–µ–º–∞ —ï–∏–¥–æ–≤–∏ –≤–æ –≤–Ω–∞—Ç—Ä–µ—à–Ω–æ—Å—Ç–∞.
            new_agent_pos = (nax, nay)
            new_box_pos = (nbx, nby)
            new_state = (new_agent_pos, new_box_pos)
            next_states.append(new_state)

    return next_states
    
def bfs(initial_state):
    queue = deque()
    queue.append(initial_state)

    visited = set()
    visited.add(initial_state)

    # –ó–∞ –¥–∞ –º–æ–∂–µ–º–µ –¥–∞ —ò–∞ —Ä–µ–∫–æ–Ω—Å—Ç—Ä—É–∏—Ä–∞–º–µ –ø–∞—Ç–µ–∫–∞—Ç–∞:
    parent = dict()  # parent[child_state] = (parent_state)

    while queue:
        current = queue.popleft()

        if is_goal(current):
            # –ì–æ –ø—Ä–æ–Ω–∞—ò–¥–æ–≤–º–µ —Ä–µ—à–µ–Ω–∏–µ—Ç–æ, —Ä–µ–∫–æ–Ω—Å—Ç—Ä—É–∏—Ä–∞–º–µ –ø–∞—Ç–µ–∫–∞
            path = []
            state = current
            while state in parent:
                path.append(state)
                state = parent[state]
            path.append(state)  # —ò–∞ –¥–æ–¥–∞–≤–∞–º–µ –∏ –ø–æ—á–µ—Ç–Ω–∞—Ç–∞ —Å–æ—Å—Ç–æ—ò–±–∞
            path.reverse()
            return path  # –≤—Ä–∞—ú–∞–º–µ –ª–∏—Å—Ç–∞ –æ–¥ —Å–æ—Å—Ç–æ—ò–±–∏

        for next_state in expand_state(current):
            if next_state not in visited:
                visited.add(next_state)
                parent[next_state] = current
                queue.append(next_state)

    return None  # –∞–∫–æ –Ω–µ–º–∞ —Ä–µ—à–µ–Ω–∏–µ
```
